```python
import pandas as pd
import matplotlib.pyplot as pit
import seaborn as sns
import scipy.stats
import numpy as np

xl = pd.ExcelFile("/content/TableauSalesData.xlsx")
SalesData = xl.parse("Orders")

#Technology highest profit/sales, Office Supplies Higher Profit Less Sales, Furniture Lowest Profit, Middle Sales
CategoryProfitSales = SalesData[["Category", "Profit", "Sales"]].groupby(by="Category").sum().sort_values(by="Sales")
display(CategoryProfitSales)

#East,West,Central,South < In order of profit/sales
JustTech = SalesData.loc[SalesData['Category']=="Technology"]
Tech_Segment = JustTech [["Segment","Profit","Sales"]]
Tech_Region = JustTech [["Region","Profit","Sales"]]
RegionTotalProfSales = Tech_Region.groupby(by="Region").sum().sort_values(by="Sales")
display(RegionTotalProfSales)

#South Indepth
TechSegRegion = JustTech[["Product Name","Region","Profit","Sales"]]
SouthTechSegment = TechSegRegion.loc[TechSegRegion["Region"]=="South"]
SouthTechSales = SouthTechSegment.groupby(by="Product Name").sum().sort_values(by="Sales",ascending=False)
display(SouthTechSales.head(5))
display(SouthTechSales.tail(5))


#Central Indepth
TechSegRegion = JustTech[["Product Name","Region","Profit","Sales"]]
CentralTechSegment = TechSegRegion.loc[TechSegRegion["Region"]=="Central"]
CentralTechSales = CentralTechSegment.groupby(by="Product Name").sum().sort_values(by="Sales",ascending=False)
display(CentralTechSales.head(5))
display(CentralTechSales.tail(5))

#West Indepth
TechSegRegion = JustTech[["Product Name","Region","Profit","Sales"]]
WestTechSegment = TechSegRegion.loc[TechSegRegion["Region"]=="West"]
WestTechSales = WestTechSegment.groupby(by="Product Name").sum().sort_values(by="Sales",ascending=False)
display(WestTechSales.head(5))
display(WestTechSales.tail(5))

#East Indepth
TechSegRegion = JustTech[["Product Name","Region","Profit","Sales"]]
EastTechSegment = TechSegRegion.loc[TechSegRegion["Region"]=="East"]
EastTechSales = EastTechSegment.groupby(by="Product Name").sum().sort_values(by="Sales",ascending=False)
display(EastTechSales.head(5))
display(EastTechSales.tail(5))

#Average Discount is 40-70%, which is good for sales but hurting profits
JustTechDiscount= JustTech[["Product Name","Discount","Sales"]]
AveDiscount = JustTechDiscount.mean()
print(AveDiscount)
JustTechAveDiscount = JustTechDiscount.groupby(by="Product Name").mean().sort_values(by="Discount",ascending=False)
display(JustTechAveDiscount.head(25))
display(JustTechAveDiscount.tail(25))


#Showing Product Name, Sales, Discount
TechProduct = JustTech[["Product Name","Sales","Discount"]]
TechTotalProfAveDiscount = TechProduct.groupby("Product Name").agg({"Sales":'sum', "Discount":"mean"})
display(TechTotalProfAveDiscount)

#Bar Chart
JustTech = SalesData.loc[SalesData["Category"]=="Technology"]
JustTechYears = JustTech.copy()
JustTechYears["Year"] = JustTechYears["Order Date"].dt.year
JustTechYearsSales = JustTechYears[["Year","Sales"]].groupby(by="Year").sum()
JustTechYearsSales = JustTechYearsSales.reset_index()
barchart1= sns.barplot(x="Year", y="Sales", data = JustTechYearsSales)
barchart1.set_title("Technology Sales by Year")
pit.show()

#PhoneSalesByRegion
JustPhones = SalesData.loc[SalesData["Sub-Category"]=="Phones"]
PhonesbyRegion = JustPhones[["Region","Product Name","Sales"]]





JustOS = SalesData.loc[SalesData['Category']=="Office Supplies"]
OS_Segment = JustOS [["Segment","Profit","Sales"]]
OS_Region = JustOS [["Region","Profit","Sales"]]
RegionTotalProfSales = OS_Region.groupby(by="Region").sum().sort_values(by="Sales")
display(RegionTotalProfSales)


#JustOSDiscount= JustOS[["Product Name","Discount","Sales"]]
JustOSDiscount = JustOS[["Product Name","Discount","Sales"]]
AvgDiscount = JustOSDiscount.mean()
JustOSAvgDiscount = JustOSDiscount.groupby(by="Product Name").mean().sort_values(by="Discount",ascending=False)
display(JustOSAvgDiscount.tail(50))


```
