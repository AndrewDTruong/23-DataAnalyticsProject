```python
#Andrew Truong
#Increasing Sales by 20%
import pandas as pd
import matplotlib.pyplot as pit
import seaborn as sns
import scipy.stats
import numpy as np
xl = pd.ExcelFile("/content/TableauSalesData.xlsx")
SalesData = xl.parse("Orders")
JustOS = SalesData.loc[SalesData["Category"]=="Office Supplies"]

#menu1 functions
def CategorySales():
  CategoryProfitSales = SalesData[["Category", "Profit", "Sales"]].groupby(by="Category").sum().sort_values(by="Sales")
  display(CategoryProfitSales)
  print("\n")
  print("By looking at the Three main Categories of Sales, we can focus on the lowest sales category which is Office Supplies")

def BarTechSalesPerYear():
  JustOSYears = JustOS.copy()
  JustOSYears["Year"] = JustOSYears["Order Date"].dt.year
  JustOSYearsSales = JustOSYears[["Year","Sales"]].groupby(by="Year").sum()
  JustOSYearsSales = JustOSYearsSales.reset_index()
  barchart1= sns.barplot(x="Year", y="Sales", data = JustOSYearsSales)
  barchart1.set_title("Office Supplies Sales by Year")
  pit.show()
  print("\n")
  print("By looking at Office Supplies Sales by Year, you can see that the sales are decreasing after 2018")

def ZeroDiscountOS():
  JustOSDiscount = JustOS[["Product Name","Discount","Sales"]]
  JustOSAvgDiscount = JustOSDiscount.groupby(by="Product Name").mean().sort_values(by="Discount",ascending=False)
  OSProf = JustOS[["Product Name", "Discount"]].groupby("Product Name").sum()
  NoDiscOS = OSProf[OSProf["Discount"] == 0.0]
  display(NoDiscOS)
  print("\n")
  print("There are 123 items with no discount in the Office Supplies Category. If we were to add discount or promotions for any of these 123 items, this will result in an increase of sales!")

def TopSalesOS():
  OSProducts = JustOS[["Product Name","Sales"]]
  OSTotalProfAveDiscount = OSProducts.groupby(by="Product Name").mean().sort_values(by="Sales",ascending=False)
  display(OSTotalProfAveDiscount.head(5))
  print("\n")
  print("Since the Letter Openers and Binding Systems have the highest sales, we can create promotions for the top selling items to increase overall sales.")

def AvgOSDiscounts():
  JustOSYear = JustOS.copy()
  JustOSYear["Year"] = JustOSYear["Order Date"].dt.year
  Years = JustOSYear.Year.unique()
  RegJustTableYear = JustOSYear[["Region","Year","Discount"]]
  for year in Years:
    OneYear = RegJustTableYear.loc[RegJustTableYear["Year"]== year]
    NoYear = OneYear[["Region","Discount"]]
    YearlyRegDiscount = NoYear.groupby(by="Region").mean()
    print("\nRegional Office Supplies discounts for the year: " + str(year))
    print(YearlyRegDiscount)
    print("*"*40)
  print("\n")
  print("By looking at the average Office Supplies Discounts by Region, you can determine why some Regions have less sales than others. ")

#Increase Profits by 10%
#All sub-categories which are sorted by the most profitable ones at the top and least profitable at the bottom
def TopSubCategories():
  subcatprofits = SalesData[['Sub-Category','Profit','Sales']]
  totalsubcatprofits = subcatprofits.groupby(by = 'Sub-Category').sum().sort_values(by ='Profit', ascending= False)
  display(totalsubcatprofits)
  print('\n')
  print('By looking at the most profitable sub-categories, we can determine which one has the highest profits out of each')

def TopTenProducts():
  productIDprofitsales = SalesData [['Product Name','Profit','Sales']].groupby(by ='Product Name').sum().sort_values(by = 'Profit',ascending= False)
  display(productIDprofitsales.head(10))
  print('\n')
  print('Comparing the top ten most profitable products to the top sub-categories can allow us to better gauge what is most profitable')

def CopierprofSalesSegment():
  justcopiers = SalesData.loc[SalesData['Sub-Category']=='Copiers']
  tablesegment = justcopiers [['Segment','Profit','Sales']]
  segtotalprofsales = tablesegment.groupby(by='Segment').sum().sort_values(by = 'Profit', ascending = False)
  display(segtotalprofsales)
  print('\n')
  print('Looking through the profits and sales by each segment for Copiers can allow us to better understand our top customers')

def CopierprofSalesRegion():
  justcopiers = SalesData.loc[SalesData['Sub-Category']=='Copiers']
  copyregion = justcopiers[['Region','Profit','Sales']]
  regiontotalprofsales = copyregion.groupby(by = 'Region').sum().sort_values(by='Profit', ascending = False)
  display(regiontotalprofsales)
  print('\n')
  print('Looking at the top sales and profits for each region for copiers can allow us to better understan the customer')

def SegmentCopierDiscounts():
  justcopiers = SalesData.loc[SalesData['Sub-Category']=='Copiers']
  justcopiersyear = justcopiers.copy()
  justcopiersyear['Year'] = justcopiersyear['Order Date'].dt.year
  Years = justcopiersyear.Year.unique()
  segjustcopiersyear = justcopiersyear[['Segment','Year','Discount']]
  for year in Years:
    oneyear = segjustcopiersyear.loc[segjustcopiersyear['Year'] == year]
    noyear = oneyear[['Segment','Discount']]
    yearlysegdiscount = noyear.groupby(by = 'Segment').mean()
    print('\nCopiers discount by customer segment for the year: ' + str(year))
    print(yearlysegdiscount)
    print('*'*40)
  print('\n')
  print('Seeing the average copier discount by customer segment for each year can allow us to better undestand what makes copiers profitable')

#Menu1_AndrewTruong
def menu_1():
  print("Welcome to Andrew's Office Supplies Data Analytics System")
  print("\n Enter 1 to see Overall Sales by Category" +
      "\n Enter 2 to see a Bar Chart containing Office Supplies Sales by Year " +
      "\n Enter 3 to see all Office Supplies items with no discounts " +
      "\n Enter 4 to see Top Sales in Office Supplies" +
      "\n Enter 5 to see the Average Office Supplies Discounts by Region for each year" +
      "\n Enter 6 to see David's Office Supplies Analytics System"
      "\n Enter 7 to exit the menu")
  choice = input("Please enter your selection here: ")
  print("\n")
  if choice == "1":
    CategorySales()
  elif choice == "2":
    BarTechSalesPerYear()
  elif choice == "3":
    ZeroDiscountOS()
  elif choice == "4":
    TopSalesOS()
  elif choice == "5":
    AvgOSDiscounts()
  elif choice == "6":
    menu_2()
  elif choice == "7":
    print("\n")
    print("Thank you for using Andrew's Office Supplies Data Analytics System")
    return None
  else:
    print("\n")
    print("You have entered an invalid option, please select from one of the options provided: ")
  menu_1()


#Menu2_DavidMoreno
def menu_2():
  print("Welcome to David's Office Supplies Data Analytics System"
      '\n Enter 1 to see the Most Profitable Sub-Categories,sorted by most to least profitable' +
        '\n Enter 2 to see the Top 10 most profitable products overall' +
        '\n Enter 3 to see the total copier profits and sales for each segment'+
        '\n Enter 4 to see the total copier profits and sales for each region'+
        '\n Enter 5 to see the average copier discount by each segment for each year'+
        "\n Enter 6 to see Andrew's Office Supplies Data Analytics System"
        '\n Enter 7 to quit\n')
  choice = input('Please enter your selection here: ')
  print('\n')
  if choice == '1':
    TopSubCategories()
    menu_2()
  elif choice == '2':
    TopTenProducts()
    menu_2()
  elif choice == '3':
    CopierprofSalesSegment()
    menu_2()
  elif choice == '4':
    CopierprofSalesRegion()
    menu_2()
  elif choice == '5':
    SegmentCopierDiscounts()
    menu_2()
  elif choice == '6':
    menu_1()
  elif choice =='7':
    print("\n")
    print("Thank you for using the David's Office Solutions Data Analytics System")
    return None
  else:
    print("\n")
    print('You have entered an invalid option, please select from one of the options provided: ')
    menu_2()

#Menu Layering
print("Welcome to Andrew and David's Office Supplies Data Analytics System")
def menu():
  print("\n Enter 1 to see Andrew's Office Supplies Data Analytics System" +
      "\n Enter 2 to see David's Office Supplies Data Analytics System " +
      "\n Enter 3 to exit ")
  choice = input("Please enter your selection here: ")
  print("\n")
  if choice == "1":
    menu_1()
  elif choice == "2":
    menu_2()
  elif choice == "3":
    print("\n")
    print("Thank you for using Office Supplies Data Analytics System")
    return None
  else:
    print("\n")
    print("You have entered an invalid option, please select from one of the options provided: ")


menu()





```
